```python
df1.columns=["ID","Cycle","OpSet1","OpSet2","OpSet3","SensorMeasure_1","SensorMeasure_2","SensorMeasure_3","SensorMeasure_4","SensorMeasure_5","SensorMeasure_6","SensorMeasure_7","SensorMeasure_8","SensorMeasure_9","SensorMeasure_10","SensorMeasure_11","SensorMeasure_12","SensorMeasure_13","SensorMeasure_14","SensorMeasure_15","SensorMeasure_16","SensorMeasure_17","SensorMeasure_18","SensorMeasure_19","SensorMeasure_20","SensorMeasure_21"]
```

```python
EOL = []
for i in df1['ID']:  # Use unique IDs to avoid redundant iterations
    last_cycle = df1[df1['ID'] == i]['Cycle'].values[-1]  # Get the last cycle for the ID
    EOL.append(last_cycle)
```

```python
df1['EOL']=EOL
```

```python
df1
```

Classification Problem Formulation

The challenge of the completion was to predict the remaining useful life of the engines by using the given sensors data and operational conditions.

But in the project, we will try to simplify that by converting it to a classification problem, where the classification will be of 3 types, ie:Good condition,Moderate condition and Warning condition

labels corresponding to each condition:

.Good condition-0

.Moderate Condition-1

.Warning condition-2


Defining the labels

Here we will define the engines condition with Life Ratio(LR) which is the ratio between Current Cycle and and the End cycle/End of Life(EOL). If LR=0, that means the component has just started its degradtion and LR=1 means,it is completely degraded.


If LR<=0.6 -Good Condition

If 0.6<LR<=0.8 -Moderate Condition

If 0.8<LR -Warning Condition



```python
# Calculating LR

df1['LR']=df1['Cycle']/df1['EOL']
```

```python
df1
```

```python
#Drop ID and EOL

df1=df1.drop(columns=['ID','EOL'])
```

```python
df1
```

```python
labels=[]
for i in range(0,len(df1)):
    if np.array(df1['LR'])[i]<=0.6:
        labels.append(0)
    elif np.array(df1['LR'])[i]<=0.8:
        labels.append(1)
    else:
        labels.append(2)
```

```python
df1['labels']=labels
df1=df1.drop(columns=['LR'])
df1
```

```python
df1.to_csv('Training_1_all_features.csv',index=None)
```

```python
df=pd.concat([df1,df2,df3,df4])
```

```python
df
```

X in Feature Matrix and y in the labels column

```python
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.10,random_state=0)
```

Training a Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred_RF=RF.predict(X_test)
cm=pd.DataFrame(confusion_matrix(y_test,y_pred_RF,normalize='true')*100)
colormap=sns.color_palette('Reds')
sns.heatmap(cm,annot=True,cmap=colormap)
accuracy_score(y_test,y_pred_RF)
```

```python
RF.get_params()
```

Performance on Training Dataset

```python
y_pred_RF=RF.predict(X_train)
cm=pd.DataFrame(confusion_matrix(y_train,y_pred_RF,normalize='true')*100)
colormap=sns.color_palette('Blues')
sns.heatmap(cm,annot=True,cmap=colormap)
accuracy_score(y_train,y_pred_RF)
```

Randomised Search with cross validation for Hyperparameter Optimization


As the previous model clearly overfits the training data,we need to do some hyperparameter tuning.For that we will do a random search with 3 fold cross validation.

```python
from sklearn.model_selection import RandomizedSearchCV
#Number of trees in random forest
n_estimators=[200,400,300]
#Maximum number of levels in tree
max_depth=[10,20,30]
#Max number of samples given to each tree
max_samples=[0.2,0.3,0.4]

#Create a random grid
random_grid= {'n_estimators':n_estimators,
              'max_depth':max_depth,
              'max_samples':max_samples}

print(random_grid)
```

```python
#Use the random grid to search for best hyperparameters
#First create the base model to tune
rf=RandomForestClassifier()
#Random search of parameters,using 3 fold cross validation
#search across 100 different combinations, and use all available cores
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,cv=3,n_jobs=-1)
#Fit the random search model
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
RF_randSer=RandomForestClassifier(n_estimators= 400, max_samples= 0.4, max_depth= 30)
RF_randSer.fit(X_train,y_train)
y_pred_RF=RF_randSer.predict(X_test)
cm=pd.DataFrame(confusion_matrix(y_test,y_pred_RF,normalize='true')*100)
colormap=sns.color_palette('Blues')
sns.heatmap(cm,annot=True,cmap=colormap)
accuracy_score(y_test,y_pred_RF)
```
