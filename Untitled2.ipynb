```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
import pandas as pd

# Load the dataset
df1 = pd.read_csv('train_FD001.txt', sep=' ', header=None)

# Drop extra empty columns (if any)
df1 = df1.dropna(axis=1, how='all')

# Dynamically assign column names
num_sensors = df1.shape[1] - 5  # Calculate the number of sensor columns
df1.columns = ['Engine_ID', 'Cycle', 'Op_Set1', 'Op_Set2', 'Op_Set3'] + [f'Sensor_{i}' for i in range(1, num_sensors + 1)]

df1

```

```python
df1= df1.sort_values(by=['Engine_ID', 'Cycle']).reset_index(drop=True)
```

# Calculate RUL
#RUL=EOL-Cycle
RUL=Remaining Useful Life
EOL=End of LIfe in Cycle


```python
EOL = []
for i in df1['Engine_ID']:  # Use unique IDs to avoid redundant iterations
    last_cycle = df1[df1['Engine_ID'] == i]['Cycle'].values[-1]  # Get the last cycle for the ID
    EOL.append(last_cycle)
```

```python
df1['EOL']=EOL
df1
```

```python
df1['RUL']=df1['EOL']-df1['Cycle']
df1=df1.drop(columns=['EOL'])
df1
```
