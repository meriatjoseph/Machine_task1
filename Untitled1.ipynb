```python
df1.columns=["ID","Cycle","OpSet1","OpSet2","OpSet3","SensorMeasure_1","SensorMeasure_2","SensorMeasure_3","SensorMeasure_4","SensorMeasure_5","SensorMeasure_6","SensorMeasure_7","SensorMeasure_8","SensorMeasure_9","SensorMeasure_10","SensorMeasure_11","SensorMeasure_12","SensorMeasure_13","SensorMeasure_14","SensorMeasure_15","SensorMeasure_16","SensorMeasure_17","SensorMeasure_18","SensorMeasure_19","SensorMeasure_20","SensorMeasure_21"]
```

```python
given_no_of_cycles=[]
for i in range(1,len(df1['ID'].unique())+1):
    given_no_of_cycles.append(len(df1[df1['ID']==i]))
    print(len(df1[df1['ID']==i]))

given_no_of_cycles=np.array(given_no_of_cycles).reshape([cycle_ran_after.shape[0],1])
```

We can see that data for 31cycles are available for engine1 in test set.Data for 40 cycles are available for engine2 and so on. Using the available data our goal is to predict RUL for each engine

```python
len(df1['ID'].unique())
```

```python
cycle_ran_after=pd.read_csv('RUL_FD001.txt', sep=' ',header=None)
cycle_ran_after = cycle_ran_after.drop([1],axis='columns')
cycle_ran_after=np.array(cycle_ran_after)
cycle_ran_after


```

```python
EOL = []
for i in df1['ID']:  # Use unique IDs to avoid redundant iterations
     EOL.append(given_no_of_cycles[i-1]+cycle_ran_after[i-1])
EOL=np.array(EOL).reshape([len(EOL),1])
```

```python
EOL.shape
```

```python
df1['EOL']=EOL
df1
```

```python
df1['LR']=df1['Cycle']/df1['EOL']
#Drop ID and EOL

df1=df1.drop(columns=['ID','EOL'])
df1
```

```python
labels=[]
for i in range(0,len(df1)):
    if np.array(df1['LR'])[i]<=0.6:
        labels.append(0)
    elif np.array(df1['LR'])[i]<=0.8:
        labels.append(1)
    else:
        labels.append(2)

df1['labels']=labels
df1=df1.drop(columns=['LR'])
df1
```

```python
df1.to_csv('text_classification_4.csv', index=None)
```
